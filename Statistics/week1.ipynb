## 1주차 - 실습 1
- ### 실습 내용
  - mean - np.mean
  - median - np.meidan
  - trim - stats.trimboth, starts.trim_mean (scipy.stats)
  - variance - np.var
  - std - np.std
  - Q - np.percentile

```python
data =[ 
    373, 477, 354, 391, 339,
    329, 405, 464, 437, 440,
    316, 379, 469, 376, 330,
    467, 436, 408, 382, 414,
    339, 367, 386, 477, 331,
    402, 384, 434, 346, 335,
    482, 420, 439, 396, 327,
    370, 391, 358, 385, 369,
    500, 355, 420, 399, 388,
    424, 303, 475, 335, 331
]
```

```python
import scipy.stats as stats
import numpy as np
mean = np.mean(data)
print(mean)
median = np.median(data)
print(median)

#trim data 0.1 from both ends
trimmed_data = stats.trimboth(data, 0.1)
#trim 0.1 from both ends -> mean
trimmed_mean = stats.trim_mean(data, 0.1)
print(trimmed_mean)
#variance
variance = np.var(data) #모분산
print(variance)
#standard deviation
std_dev = np.std(data) #모표준편차
print(std_dev)
print(min(data),max(data))
#Q1,Q3, IQR
Q1 = np.percentile(data, 25)
Q2 = np.percentile(data, 50) #median
Q3 = np.percentile(data, 75)
print(Q1,Q2,Q3)
#IQR
IQR = Q3 - Q1 #interquartile range
print(IQR)
#sd/mean  : coefficient of variation
cv = std_dev/mean
print(cv)

#plot histogram
import matplotlib.pyplot as plt
plt.hist(data, bins=4, alpha=0.5) #alpha 투명도
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats

#Q1 평균
mean = np.mean(data)

#Q2 중앙값
median = np.median(data)

#Q3 10% 절사평균
trimmed_mean_10 = stats.trim_mean(data,0.1)

#Q4 분산
variance = np.var(data)

#Q5 표준편차
std_dev = np.std(data)

#Q6~7 최소값,최대값
min_value, max_value = min(data), max(data)

#Q8 범위
range_value = max_value - min_value

#Q9~11 1사분위, 2사분위, 3사분위
Q1,Q2,Q3 = np.percentile(data, [25,50,75])

#12 사분위수 범위(IQR)
IQR = Q3-Q1

#13 변동계수
cv = std_dev/mean

#14 히스토그램을 작성하고 분포의 모양에 대하여 설명
plt.hist(data, bins=10, alpha=0.5, edgecolor='black')
plt.show()
#오른쪽으로 치우친 분포

#15 정규성 검정
from scipy.stats import shapiro
stat, p = shapiro(data)
print(stat, p)
#W = 0.89, p-value = 0.0034 -> 기각 -> 정규성을 만족하지 않음
```
